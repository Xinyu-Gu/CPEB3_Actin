```python
import matplotlib.pyplot as plt
import numpy as np

#cn=np.loadtxt('tmp.dat')
c=np.loadtxt('mean.log')
e=np.loadtxt('../../peptide.raw')
cn=np.transpose(c)[-1]

print(cn)

pp=-(cn - np.max(cn))
ep=e-np.min(e)
    

    
plt.plot(pp/np.max(pp))
plt.plot(ep/np.max(ep))

plt.xlabel('Peptide Index', fontsize=16)
plt.legend(('|binding Energy|','Peptide Array Intensity'))#, loc='upper left')
plt.savefig("expt-inter.png")
```
